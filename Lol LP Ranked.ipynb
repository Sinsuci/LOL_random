```python
import random
import matplotlib.pyplot as plt

wr = 66
lp_aim = 150
echantillon_num = 100
max_game = 1000
lp_gain = 15
lp_loss = 15


echantillon = 0
moyenne = 0
history_all = []


while echantillon < echantillon_num :
    lp = 0
    gm = 0
    echantillon = echantillon + 1
    history = []

    while lp < lp_aim :
        gm = gm + 1
        if gm > max_game :
            break
        rand = int(100*random.random())
        if wr < rand :
            lp = lp - lp_loss
        else :
            lp = lp + lp_gain
        history.append(lp)
    
    moyenne = moyenne + gm
    history_all.append(history)


moyenne = int(moyenne/echantillon_num)


for i in history_all :
    plt.plot(i)


esperance = int(lp_aim / (lp_gain * wr / 100 - lp_loss * (1-(wr / 100))))

print("esperance : ",esperance,'\n',"moyenne de game : ",moyenne)

plt.xlabel('Games')
plt.ylabel('LP')
plt.show()
```
